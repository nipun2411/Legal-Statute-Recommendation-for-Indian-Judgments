```python
# quick check
import torch
print("torch:", torch.__version__, "cuda_available:", torch.cuda.is_available())
```

```python
# 🚀 Install faster-whisper
!pip install -U faster-whisper

# --- Python code ---
from faster_whisper import WhisperModel

# Load a smaller model for Hindi (tiny/base/small are faster and lighter than large)
# You can swap "small" with "tiny" or "base" if GPU memory is low
model = WhisperModel("small", device="cuda", compute_type="float16")

# Path to your Hindi audio file in Drive
audio_path = "/content/drive/MyDrive/stt_project/audio/hindi.mp3"

# Transcribe
segments, info = model.transcribe(audio_path, language="hi")

# Print combined transcript
transcript = " ".join([seg.text for seg in segments])
print("Hindi transcript:\n", transcript)
```

```python
# 🚀 Install faster-whisper
!pip install -U faster-whisper

import os
from faster_whisper import WhisperModel

# Load a model (try "medium"; if GPU runs out of memory, switch to "small" or "base")
model = WhisperModel("medium", device="cuda", compute_type="float16")

# Path to your Marathi audio file in Drive
audio_path = "/content/drive/MyDrive/stt_project/audio/marathi.mp3"

# ✅ Check if the file exists
if not os.path.exists(audio_path):
    print(f"❌ File not found at: {audio_path}")
    print("Tip: Run this to check available files:")
    print("!ls '/content/drive/MyDrive/stt_project/audio/'")
else:
    # Transcribe (faster-whisper auto-chunks long audio, no warnings needed)
    segments, info = model.transcribe(audio_path, language="mr")

    # Join all segments into one transcript string
    transcript = " ".join([seg.text for seg in segments])
    print("✅ Marathi transcript:\n", transcript)
```
